```python
import pandas as pd
df = pd.read_csv('FullYearTiming.csv')
```

```python
df.head()
```

```python
#set the time 24 hour format
def change_format_to_24hour(x):
    a = x.split(':')
    a[0] = str(int(a[0])+12)
    x = ":".join(a)
    return x
prayers_to_in_24h = ['DHUHR prayer','ASR prayer','M/RIB prayer','ISHA prayer']
for i in prayers_to_in_24h:
    df[str(i)] = df[str(i)].apply(change_format_to_24hour)
```

```python
df.head()
```

```python
masjid_name = 'Masjid'
import datetime as dt
#get the today's date in the format of 01-Jan
today = dt.datetime.today().strftime('%d-%b')
#get the today's date in the US format
today_date_fmt = dt.datetime.today().strftime('%d-%b-%Y')


today_date_fmt
```

```python
#get the row of today's date from the dataframe
today_row = df[df['GregorianMonthDay'] == today]
today_row
#get the row of tomorrow's date from the dataframe
tomorrow = dt.datetime.today() + dt.timedelta(days=1)
tomorrow_date_fmt = tomorrow.strftime('%d-%b-%Y')
tomorrow = tomorrow.strftime('%d-%b')
tomorrow_row = df[df['GregorianMonthDay'] == tomorrow]
tomorrow_row
tomorrow_date_fmt
```
